```python
from fredapi import Fred
import pandas as pd

# Replace 'your_api_key' with your actual FRED API key
fred = Fred(api_key='e7d91fd4ac05e8ae87fa956d551f861f')
```

```python
# Download monthly industrial production data from January 1965 to present
gdp = fred.get_series('GDP', observation_start='1965-01-01')

# Convert the series to a DataFrame
gdp_df = gdp.to_frame(name='GDP')
print(gdp_df)
```

```python
# Graph the data
import matplotlib.pyplot as plt
gdp_df.plot(title='GDP', legend=None)
plt.show()
```
